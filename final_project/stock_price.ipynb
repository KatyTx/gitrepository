```python
#import libraries
import math
import pandas_datareader as web
import numpy as np
import pandas as pd
```

```python
# read csv for top list
stock_list = pd.read_csv('sandp100.csv')
stock_list.head()
```

```python
#extract stock symbols
tickers = stock_list['Symbol']
tickers
```

```python
#replace some values that are presented different in yahoo
stock_list['Symbol'] = stock_list['Symbol'].replace(['BRK.B'],'BRK-B')
```

```python
#extract symbols into separate list
ticker_list = stock_list['Symbol'].tolist()
```

```python
ticker_list
```

```python
%%time
# get one stock quote to verify request is working
df1 = web.DataReader('AAPL', data_source = 'yahoo', start= '2020-01-01', end= '2020-12-31')
df1
```

```python
%%time
#run complete ticker list 
p = web.DataReader(ticker_list, 'yahoo', start= '2020-01-01', end= '2020-12-31')
```

```python
#reset index
res = p.stack().reset_index()
```

```python
#adjust columns and index
res.columns.names=[None]
res.index.names = ['ID']
```

```python
print(res)
```

```python
#store web requests into original data frame
stock_data_df_org = pd.DataFrame(res) 
stock_data_df_org.head()
```

```python
print(stock_data_df_org.dtypes)
print(stock_data_df_org.columns)
```

```python
#store original data frame into working data frame
stock_data_df= stock_data_df_org
stock_data_df
```

```python
#adjust pricing formating to 2 decimal points
pd.options.display.float_format = '{:,.2f}'.format
print(stock_data_df)
print(stock_data_df.head())
```

```python
#adjust the volume column to remove extra decimals
stock_data_df['Volume']=stock_data_df['Volume'].astype(int).apply(lambda x : "{:,}".format(x))
stock_data_df.head()
```

```python
stock_data_df= stock_data_df.rename(columns={'Symbols': 'Symbol', 'Adj Close': 'Adj_Close'})
stock_data_df.head()
```

```python
%%time
#create dataframe of each stock symbol stock price request from yahoo from 1/1/2020 thru 12/31/20
#alternative method testing one data set
dict_of_tickers = {ticker: web.DataReader(ticker_list[:1], data_source = 'yahoo', start= '2020-01-01', end= '2020-12-31') \
                   for ticker in ticker_list}
```
